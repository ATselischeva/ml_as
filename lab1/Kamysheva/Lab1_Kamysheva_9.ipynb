### 2. Вычислите корреляцию Пирсона и Спирмена между признаками Age и G. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.

```python
data_copy = data.dropna(subset=['G', 'Age'])
pearson_cor, pearson_pvalue = stats.pearsonr(data_copy['G'], data_copy['Age'])
spearman_cor, spearman_pvalue = stats.spearmanr(data_copy['G'], data_copy['Age'])

print('Pearson: correlation     pvalue')
print('        ',pearson_cor,'', pearson_pvalue)

print('Spearman: correlation     pvalue')
print('        ',spearman_cor,'', spearman_pvalue)

print('Так как значения корреляции близки к нулю, это говорит о том, что связь признаков практически отсутствует')

print('AGE')
sns.countplot(data=data, y='Age')
plt.show()

print('GAMES')
sns.countplot(data=data, y='G')
plt.show()
```

### 8. Определите самое популярное имя среди игроков.

```python
from collections import Counter

names = data.Player.apply(lambda x: x.split(' ')[0] if type(x)==str else x)
sonames = data.Player.apply(lambda x: x.split(' ')[1] if type(x)==str else x)

players = {(names[i], sonames[i]) for i in range(len(names))}
un_names = tuple(player[0] for player in players)
print("Most popular name:", max(Counter(un_names).items(), key=lambda x: x[1])[0])
```
